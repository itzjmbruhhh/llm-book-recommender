## Imports

```python
import pandas as pd
import kagglehub
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Loading the DataSet via KaggleHub

```python
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

## Heatmap to find patterns

```python
# Heatmap to find patterns with missing data
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

## Looking for patterns


```python
# Creating numerical representation column for description column. 1 if missing and 0 if not
books["missing_description"] = np.where(books["description"].isna(), 1, 0)

# Calculate book age based on publication_year column
books["age_of_book"] = 2026 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

# Create a correlation matrix
correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label" : "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

Values closer to `-1` or `0` indicate little to no correlation, while values closer to `1` indicate a stronger positive correlation.

The correlation matrix shows that there is not much correlation between missing values across the other columns of interest.

```python
# Since filling up missing values for description can be quite tedious, it is safe for us to delete it
# But first we must find out how many books we are actually going to remove

# view books that have missing values
books[( books["description"].isna()) |
        (books["num_pages"].isna()) |
        (books["published_year"].isna()) |
        (books["average_rating"].isna())
]
```

```python
g# Assign books that do are not missing across the selected features in a variable
books_missing = books[~( books["description"].isna()) &
        ~(books["num_pages"].isna()) &
        ~(books["published_year"].isna()) &
        ~(books["average_rating"].isna())
]

books_missing
```

```python
# Checking number of categories based on descending counts of unique values
books_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)

# We can see a 'long tail' problem when looking at the category's count chart
```

```python
# Removing useless/short 'description' entries, by measuring its length

# Add another column that counts description length with string splitting
books_missing["words_in_description"] = books_missing["description"].str.split().str.len()
```

```python
books_missing
```

### Looking for patterns, in between lengths of descriptions

```python
books_missing.loc[books_missing["words_in_description"].between(1, 4), "description"]
# Descriptions with length ranging from 1 to 4 are pretty much useless and does not give any good inputs
```

```python
books_missing.loc[books_missing["words_in_description"].between(5, 14), "description"]
# Descriptions with length ranging from 5 to 14 are pretty much useless and does not give any good inputs
```

```python
books_missing.loc[books_missing["words_in_description"].between(15, 24), "description"]
# Descriptions with length ranging from 15 to 24 starts to give of meaningful results but not enough
```

```python
books_missing.loc[books_missing["words_in_description"].between(25, 34), "description"]
# Descriptions with length ranging from 25 to 34 gives us sufficient information
```

```python
# Filter out books that have less than 25 length of description
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
books_missing_25_words
```

### Making 'subtitle' column useable

```python
title = books_missing_25_words["title"].fillna("").astype(str)
subtitle = books_missing_25_words["subtitle"].fillna("").astype(str)

books_missing_25_words["title_and_subtitle"] = np.where(
    books_missing_25_words["subtitle"].isna(), # if subtitles are missing, then just use the title
    title,
    title + ": " + subtitle # if title and subtitle both exist, then aggregate it as one
)
```

```python
books_missing_25_words
```

### Separate description with unqiue identifiers, for this instance we can use ISBN as identifiers as they are already unique

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
books_missing_25_words
```

### Dropping uneeded columns and save it as csv

```python
(
    books_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
